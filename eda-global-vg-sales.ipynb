Rank: Ranking of the game based on global sales.

Name: Name of the game.

Platform: Platform the game was released on.

Year: Year the game was released.

Genre: Genre of the game.

Publisher: Publisher of the game.

NA_Sales: Sales of the game in North America.

EU_Sales: Sales of the game in Europe.

JP_Sales: Sales of the game in Japan.

Other_Sales: Sales of the game in other regions.

Global_Sales: Total sales of the game worldwide.

```r
is_any_na = function(x){
  return(any(is.na(x)))
}
```

```r
genres = unique(df$Genre)
years = unique(df$Year)
dfA = data.frame(Genre = character(), Year = integer(), Total_Sales = double())
for (genre in genres) {
  for (year in years) {
    total_sales = sum(df$Global_Sales[df$Genre == genre & df$Year == year])
    dfA = rbind(dfA, data.frame(Genre = genre, Year = year, Total_Sales = total_sales))
  }
}
```

```r
ggplot(dfA, aes(x = Year, y = Total_Sales, color = Genre)) +
  geom_point() +
  labs(title = "Global Sales of Each Genre per Year (FIG.9)",
       x = "Year",
       y = "Global Sales",
       color = "Genre") +
  theme_minimal()+
     theme(axis.text.x = element_text(angle = 50, hjust = 1))+
     scale_color_manual(values = c("Action" = "Purple",
   "Adventure" = "Blue",
   "Fighting" = "Hot Pink",
   "Misc" = "Green",
   "Platform" = "Yellow",
   "Puzzle" = "Sky blue",
   "Racing" = "Black",
   "Role-Playing"= "Brown",
   "Shooter" = "Red",
   "Simulation" = "orange",
   "Sports"= "Pink",
   "Strategy" = "Dark Green"))
```
